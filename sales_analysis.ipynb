```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(12,6))
df.groupby('year_month')[['sales','profit']].sum().plot(kind='line', marker='o')
plt.title("Sales and Profit Over Time")
plt.ylabel("Amount")
plt.show()

# 2. Sales by Category
plt.figure(figsize=(10,6))
sns.barplot(x="category", y="sales", data=df, estimator=sum, ci=None)
plt.title("Total Sales by Category")
plt.show()

# 3. Profit Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['profit'], bins=50, kde=True)
plt.title("Profit Distribution")
plt.show()

# 4. Sales vs Profit Scatter
plt.figure(figsize=(10,6))
sns.scatterplot(x="sales", y="profit", hue="segment", data=df, alpha=0.7)
plt.title("Sales vs Profit by Segment")
plt.show()

# 5. Orders by Segment
plt.figure(figsize=(8,6))
sns.countplot(x="segment", data=df)
plt.title("Number of Orders by Segment")
plt.show()

# 6. Delivery Days by Ship Mode
plt.figure(figsize=(10,6))
sns.boxplot(x="ship_mode", y="days_between_delivery", data=df)
plt.title("Delivery Days by Ship Mode")
plt.show()
```

```python
sns.set_theme(style="whitegrid")

# Convert year_month to datetime
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y-%m')

# Aggregate monthly sales
monthly_sales = df.groupby('year_month')['sales'].sum().reset_index()

# Calculate monthly growth rate (percentage)
monthly_sales['growth_rate'] = monthly_sales['sales'].pct_change() * 100

# --- Monthly Sales Trend with Growth Rate ---
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_sales, x='year_month', y='sales', marker='o', linewidth=2, color="#4C72B0")

# Add growth rate annotations
for i in range(1, len(monthly_sales)):
    plt.text(
        x=monthly_sales['year_month'][i],
        y=monthly_sales['sales'][i] + (monthly_sales['sales'].max() * 0.02),  # shift a bit upward
        s=f"{monthly_sales['growth_rate'][i]:.1f}%",
        ha='center',
        fontsize=9,
        color='darkgreen' if monthly_sales['growth_rate'][i] > 0 else 'red'
    )

# Styling
plt.title("📈 Monthly Sales Trend with Growth Rate", fontsize=14, weight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# --- Number of Sales per Month ---
plt.figure(figsize=(14, 6))
sns.histplot(
    data=df,
    x='year_month',
    bins=len(df['year_month'].unique()),
    discrete=True,
    color="#04651B",
    edgecolor="black"
)

# Styling
plt.title("🛒 Number of Sales Over Time", fontsize=14, weight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
group=df.groupby('category').agg(tsales=('sales','sum'), total_order=('order_id','count'),tprofit=('profit','sum'))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 6))  # 1 row, 3 columns

# Pie chart - Total Sales
axes[0].pie(group['tsales'], labels=group.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title("Sales Distribution")

# Pie chart - Total Profit
axes[1].pie(group['tprofit'], labels=group.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title("Profit Distribution")

# Pie chart - Total Orders
axes[2].pie(group['total_order'], labels=group.index, autopct='%1.1f%%', startangle=90)
axes[2].set_title("Orders Distribution")

plt.suptitle("Category-wise Distribution Overview", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# --- Sales by Category ---
sns.barplot(
    data=df, x='category', y='sales', estimator=np.sum, hue='payment_mode',
    ax=axes[0, 0]
)
axes[0, 0].set_title("Total Sales by Category & Payment Mode")
axes[0, 0].set_xlabel("Category")
axes[0, 0].set_ylabel("Total Sales")

# --- Profit by Category ---
sns.barplot(
    data=df, x='category', y='profit', estimator=np.sum, hue='payment_mode',
    ax=axes[0, 1]
)
axes[0, 1].set_title("Total Profit by Category & Payment Mode")
axes[0, 1].set_xlabel("Category")
axes[0, 1].set_ylabel("Total Profit")

# --- Quantity by Category ---
sns.barplot(
    data=df, x='category', y='quantity', estimator=np.sum, hue='payment_mode',
    ax=axes[1, 0]
)
axes[1, 0].set_title("Total Quantity by Category & Payment Mode")
axes[1, 0].set_xlabel("Category")
axes[1, 0].set_ylabel("Total Quantity")

# Remove empty subplot (bottom-right)
axes[1, 1].axis("off")

# Adjust layout
plt.suptitle("Category-wise Analysis by Payment Mode", fontsize=18, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.96])  # leave space for main title

# Move legend outside (so it doesn’t clutter)
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper right", bbox_to_anchor=(1.1, 0.95))

plt.show()

```
